### Data

```python
graph = nx.karate_club_graph()
```

```python
sense_feat_dict, sense_features = get_sense_features(graph, ppr_flag = 'std')

uncorrelated_feats = ['Degree',
                    'Clustering Coefficient',
                    'Personalized Page Rank - Standard Deviation',
                    'Average Neighbor Degree',
                    'Average Neighbor Clustering',
                    'Eccentricity',
                    'Katz Centrality']

if 'Eccentricity' not in sense_feat_dict:
    uncorrelated_feats = ['Degree',
                    'Clustering Coefficient',
                    'Personalized Page Rank - Standard Deviation',
                    'Average Neighbor Degree',
                    'Average Neighbor Clustering',
                    'Katz Centrality']

sense_features = sense_features[:, [list(sense_feat_dict).index(feat) for feat in uncorrelated_feats]]
sense_feat_dict = {feat : idx for idx, feat in enumerate(uncorrelated_feats)}
```

### DGI

```python
from DGI.models import DGI, LogReg
from DGI.utils import process
```

```python
dgi = DGIEmbedding(graph = graph, 
           embed_dim = 16, 
           feature_matrix = sense_features, 
           use_xm = True, 
           ortho_ = 10, 
           sparse_ = 1, 
           batch_size = 1, 
           model_name = '_')
embed = dgi.get_embedding()
embed = (embed - np.min(embed)) / np.ptp(embed)
feature_dict = find_feature_membership(input_embed = embed,
                                                    embed_name = 'DGI',
                                                    sense_features = sense_features,
                                                    sense_feat_dict = sense_feat_dict,
                                                    top_k = 8,
                                                    solver = 'nmf')

explanation = get_explanations(embed, sense_features)
```

```python
node = 0
plt.figure(figsize = (10, 5), dpi = 300)
plt.imshow(explanation[node], aspect = 'auto', cmap = 'Blues')
plt.xticks(list(range(len(sense_feat_dict))), list(sense_feat_dict), rotation = 90)
plt.colorbar()
plt.show()
```

### GMI

```python
from GMI_.models import GMI, LogReg
from GMI_.utils import process
```

```python
gmi = GMIEmbedding(graph = graph, 
           embed_dim = 16, 
           feature_matrix = sense_features, 
           use_xm = True, 
           ortho_ = 1, 
           sparse_ = 10, 
           batch_size = 1, 
           model_name = '_', 
           epoch_flag = 250)
embed = gmi.get_embedding()
embed = (embed - np.min(embed)) / np.ptp(embed)
feature_dict = find_feature_membership(input_embed = embed,
                                                    embed_name = 'GMI+XM',
                                                    sense_features = sense_features,
                                                    sense_feat_dict = sense_feat_dict,
                                                    top_k = 8,
                                                    solver = 'nmf')
explanation = get_explanations(embed, sense_features)
```

```python
node = 0
plt.figure(figsize = (10, 5), dpi = 300)
plt.imshow(explanation[node], aspect = 'auto', cmap = 'Blues')
plt.xticks(list(range(len(sense_feat_dict))), list(sense_feat_dict), rotation = 90)
plt.colorbar()
plt.show()
```

### SDNE

```python
# Standard SDNE
sdne = SDNE_plus(graph, 
                  hidden_size = [32, 16], 
                  lr = 1e-3,
                  sense_features = sense_features,
                  alpha = 0.1,
                  beta = 10, 
                  gamma = 0, 
                  delta = 0)
_ = sdne.train(epochs = 200, batch_size = 1024)

# SDNE+XM
sdne_plus = SDNE_plus(graph, 
                      hidden_size = [32, 16], 
                      lr = 1e-4,
                      sense_features = sense_features.astype(np.float32),
                      alpha = 1, 
                      beta = 1, 
                      gamma = 10, 
                      delta = 10)

sdne_plus.model.set_weights(sdne.model.get_weights())
history = sdne_plus.train(epochs = 400, batch_size = 1024)
e = sdne_plus.get_embeddings()
embed_plus = np.array([e[node_name] for node_name in graph.nodes()])
embed_plus = (embed_plus - np.min(embed_plus)) / np.ptp(embed_plus)


feature_dict_plus = find_feature_membership(input_embed = embed_plus,
                                                    embed_name = 'SDNE+XM',
                                                    sense_features = sense_features,
                                                    sense_feat_dict = sense_feat_dict,
                                                    top_k = 8,
                                                    solver = 'nmf')
explanation = get_explanations(embed_plus, sense_features)
```

```python
node = 0
plt.figure(figsize = (10, 5), dpi = 300)
plt.imshow(explanation[node], aspect = 'auto', cmap = 'Blues')
plt.xticks(list(range(len(sense_feat_dict))), list(sense_feat_dict), rotation = 90)
plt.colorbar()
plt.show()
```

### LINE

```python
# Standard LINE
line = LINE(graph, 
        embedding_size = 16,
        sense_features = sense_features,
        alpha = 0.1, 
        ortho = 0, 
        sparse = 0,
        learning_rate =  1e-3,
        order = 'second', 
        batch_size = 1024)

_ = line.train(epochs = 15)

# LINE+XM
line_plus = LINE(graph, 
                embedding_size = 16,
                sense_features = sense_features,
                alpha = 1, 
                ortho = 100, 
                sparse = 10,
                learning_rate =  5e-4,
                order = 'second', 
                batch_size = 1024)

_ = line_plus.train(epochs = 50)

e = line_plus.get_embeddings()
embed_plus = np.array([e[node_name] for node_name in graph.nodes()])
embed_plus = (embed_plus - np.min(embed_plus)) / np.ptp(embed_plus)

feature_dict_plus = find_feature_membership(input_embed = embed_plus,
                                                    embed_name = 'LINE+XM',
                                                    sense_features = sense_features,
                                                    sense_feat_dict = sense_feat_dict,
                                                    top_k = 8,
                                                    solver = 'nmf')

explanation = get_explanations(embed_plus, sense_features)
```

```python
node = 0
plt.figure(figsize = (10, 5), dpi = 300)
plt.imshow(explanation[node], aspect = 'auto', cmap = 'Blues')
plt.xticks(list(range(len(sense_feat_dict))), list(sense_feat_dict), rotation = 90)
plt.colorbar()
plt.show()
```
